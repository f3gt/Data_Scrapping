```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url ='https://www.imdb.com/chart/top/'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'
}
```

```python
page = requests.get(url,headers=headers)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
# list to store data

movies_link=[]
movie_name = []
movies_vote = []
movies_writer = []
movies_director = []
movies_star= []
movies_ratin = []
```

```python
sec_data = soup.find('ul', class_ ="ipc-metadata-list ipc-metadata-list--dividers-between sc-e22973a9-0 khSCXM compact-list-view ipc-metadata-list--base")
```

```python
movie_data = sec_data.find_all('a', class_="ipc-title-link-wrapper")
```

```python
for a in movie_data:
    extrec = a.get('href')
    final_link = f"https://www.imdb.com{extrec}"
    print (final_link)
    movies_link.append(final_link)
```

```python
for link in movies_link:

    url = link
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'
    }

    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.text, 'html.parser')

    # Movie Name
    try:
        name = soup.find('span', class_="hero__primary-text")
        name_text = name.get_text() if name else ""
    except:
        name_text = ""
    print(name_text)
    movie_name.append(name_text)

    # Rating
    try:
        rating = soup.find('div', class_="sc-d541859f-2 kxphVf")
        rating_text = rating.get_text() if rating else ""
    except:
        rating_text = ""
    print(rating_text)
    movies_ratin.append(rating_text)

    # Votes
    try:
        votes = soup.find('div', class_="sc-d541859f-3 dwhNqC")
        vote_text = votes.get_text() if votes else ""
    except:
        vote_text = ""
    print(vote_text)
    movies_vote.append(vote_text)

    # Director
    try:
        directer = soup.find('div', class_="ipc-metadata-list-item__content-container")
        director_text = directer.get_text(strip=True) if directer else ""
    except:
        director_text = ""
    print(director_text)
    movies_director.append(director_text)

    # Writers
    try:
        writers_section = soup.find('li', {'data-testid': 'title-pc-principal-credit'})
        writer_names = ""
        if writers_section and 'Writers' in writers_section.get_text():
            writer_tags = writers_section.find_all('a')
            writer_names = [tag.get_text(strip=True) for tag in writer_tags]
        movies_writer.append(writer_names)
        print("Writer names:", writer_names)
    except:
        movies_writer.append("")
        print("Writers section not found.")

    # Stars
    try:
        st = soup.find('div', class_="sc-bf57f3f2-2 dvWMNJ")
        star = st.find_all('li', {'data-testid': 'title-pc-principal-credit'})[2] if st else None

        if star and 'Stars' in star.get_text():
            Stars = star.find_all('a')[1:]
            star_name = [tag.get_text(strip=True) for tag in Stars]
        else:
            star_name = []
        movies_star.append(star_name)
        print("Stars Name:", star_name)
    except:
        movies_star.append("")
        print("Stars section not found.")
```

```python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({
    'Name': movie_name,
    'Link': movies_link,
    'Rating': movies_ratin,
    'Votes': movies_vote,
    'Writers': movies_writer,
    'Directors': movies_director,
    'Stars': movies_star
})

# Display the DataFrame
print(df.head())
```
